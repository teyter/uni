```python
#BYRJA -- Keyrið til að frumstilla numpy o.fl.
import numpy as np, numpy.random as npr
import matplotlib.pyplot as plt
import scipy.stats as st, scipy.optimize as opt
plt.rc('axes', axisbelow=True)
%matplotlib inline
np.set_printoptions(precision=2, floatmode='fixed', suppress=True)

def hlaup_medal(x):
    """skilar n-vigri með 15-staka hlaupandi meðaltali n-vigursins x"""
    n = len(x)
    y = np.zeros(n)
    for i in range(n):
        y[i] = x[max(i-7,0):min(i+7,n)].mean()
    return y
```

## S5.A1 Hlaupandi meðaltal
Hér vinnum við áfram með hitann í Stykkishólmi, sbr. fyrsta dæmið í T5. Lesið `stykk.txt` inn í tvo vigra með ártali og meðalhita ársins. Í a-lið á að nota fallið `hlaup_meðal` sem skilgreint er í byrjunarreit en það reiknar (í okkar tilviki) 15 ára hlaupandi meðaltal, þannig að y[i] verður meðaltalið af hitastigi ársins ár[i] og næstu 7 ára á undan og eftir.

**a)** Teiknið hitaröðina og 15-ára hlaupandi meðaltal inn á sömu mynd<br>
**b)** Lesið um *label* og *legend* í [grein 2.3.3 í fyrirlestrarnótum](https://notendur.hi.is/~jonasson/strei/kafli02.html#stjorn-utlits-plot-og-hist-mynda) og setjið í framhaldi hæfilegt *legend* inn á myndina<br>
**c)** Finnið þau tímabil sem hlaupandi meðaltal hitans er hærra en 3.5°C<br>

Í c-lið er hægt nota mynd, eða lesa svarið út úr útskrift með `print`, og svo væri líka hægt að nota fallið `np.where` sem skilar staðsetningum staka sem uppfylla ákveðið skilyrði – skoðið hjálpina.

```python
# a
(ar, hiti) = np.loadtxt("stykk.txt").T
y = hlaup_medal(hiti)
plt.figure(figsize=(20,8))
plt.plot(ar,hiti,label="Ársmeðalhiti")
plt.plot(ar,y,label="Hlaupandi meðtaltal")
plt.grid()
plt.title("Ársmeðalhiti á Stykkishólmi 1830-2018")
# b c
z = y > 3.5
plt.scatter(ar[z],y[z],color='green', label='meira en 3.5')
plt.legend()
plt.xlabel("Ártal")
plt.ylabel("Hitastig (C°)")
#ar[z]
#z = np.zeros(len(y),np.int)
#z[y > 3.5] = 1
#print(z)
```

## S5.B1  Stigull Rosenbrocksfallsins
   **a)** Reiknið (á blaði) stigul falls Rosenbrocks:

   $$
      f(x, y) = (1 - x)^2 + 100(y - x^2)^2
   $$

   **b)** Reiknið í framhaldi $\nabla f(-1,1)$, $\nabla f(0,1)$ og
      $\nabla f(1,1)$. Athugið hvort niðurstöðurnar rími
      við eftirfarandi hæðarlínumynd af fallinu, t.d. með því að
      rissa upp á blaðið stiglana þrjá og hæðarlínur í grennd við þá 
      (styttið þá samt svo þeir komist fyrir á blaðinu). Þið gætuð 
      líka prentað út hæðarlínumyndina og teiknað stiglana inn á útprentið.

```python
#TEIKNING ROSENBROCKSFALLS
def ros(x):
    a = (1 - x[0])**2
    b = (x[1] - x[0]**2)
    return a + 100*b**2

plt.figure(figsize=(7,7))
levels = np.linspace(0,26,14)**2; levels[0] = 0.3
X = np.linspace(-1.5, 1.5, 200)
Y = np.linspace(-0.5, 2.5, 200)
Z = np.array([[ros([x,y]) for x in X] for y in Y])
plt.gca().set(aspect='equal')
c = plt.contour(X, Y, Z, levels)
plt.clabel(c, fmt='%.1f', colors = 'k')
plt.grid()
```
